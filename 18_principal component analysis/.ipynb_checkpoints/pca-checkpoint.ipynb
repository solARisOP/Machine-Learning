```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_digits

digits = load_digits()
```

```python
digits.keys()
```

```python
df = pd.DataFrame(digits.data, columns = digits.feature_names)
df
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
plt.gray()

plt.matshow(digits.data[9].reshape(8,8))
```

```python
X = df
y = digits.target
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_tranform = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split as tts
X_train, X_test, y_train, y_test = tts(X_tranform, y, test_size = 0.2, random_state = 30)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
#this is used for dimensionality reduction as we can have n no. of dimensions we can perform pca on it and can reduce them
from sklearn.decomposition import PCA 
pca = PCA(0.95) # we can give n_components value or this decimal value below 1 , n_components mean the most important
                # and reduced no. of components we wanna pick for the model trainning and predictions but this decimal value means
                # the percent of important data we wanna use to make predcitiion 0.95 doesnot mean it will pick 95% imp columns
                # but it pick 95% imp data and make new columns with it that will help us train and make prediction with reduced columns
X_pca = pca.fit_transform(X)
X_pca.shape
```

```python
pca.n_components_ # all new columns that the pca has formed
```

```python
pca.explained_variance_ratio_
```

```python
X_train_pca, X_test_pca, y_train_pca, y_test_pca = tts(X_pca, y, test_size = 0.2, random_state = 30)
```

```python
model = LogisticRegression(max_iter = 3000)
model.fit(X_train_pca, y_train_pca)
model.score(X_test_pca, y_test_pca)
```
