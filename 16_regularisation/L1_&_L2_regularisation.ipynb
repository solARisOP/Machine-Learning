Regularisations make underfitting and overfitting model more balanced they improve accuracy of the model

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Melbourne_housing_FULL.csv')
```

```python
df
```

```python
df.nunique()
```

```python
df.shape
```

```python
cols_to_use = ['Suburb', 'Rooms', 'Type', 'Method', 'SellerG', 'Regionname', 'Propertycount', 
               'Distance', 'CouncilArea', 'Bedroom2', 'Bathroom', 'Car', 'Landsize', 'BuildingArea', 'Price']
df = df[cols_to_use]
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
cols_zero = ['Propertycount', 'Car', 'Bedroom2', 'Bathroom', 'Distance']
df[cols_zero] = df[cols_zero].fillna(0)
```

```python
df['Landsize'] = df['Landsize'].fillna(df['Landsize'].mean())
```

```python
df['BuildingArea'] = df['BuildingArea'].fillna(df['BuildingArea'].mean())
```

```python
df.dropna(inplace = True)
```

```python
df
```

```python
df = pd.get_dummies(df, drop_first = True)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split as tts
X_train, X_test, y_train, y_test = tts(df.drop(['Price'], axis = 1), df.Price, test_size = 0.3, random_state= 2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
model.score(X_train, y_train) # we can see both samples of the datasets have an accuracy of huge diff which is a clear case of overfitting, so in below we will apply regularisations to make it balanced
```

```python
from sklearn.linear_model import Lasso # L1 regularisation
lasso_model = Lasso(alpha = 50, max_iter = 100, tol = 0.1)
lasso_model.fit(X_train, y_train)
lasso_model.score(X_test, y_test)
```

```python
lasso_model.score(X_train, y_train)
```

```python
from sklearn.linear_model import Ridge # L2 regularisation
ridge_model = Ridge(alpha = 50, max_iter = 100, tol = 0.1)
ridge_model.fit(X_train, y_train)
ridge_model.score(X_test, y_test)
```

```python
ridge_model.score(X_train, y_train)
```
