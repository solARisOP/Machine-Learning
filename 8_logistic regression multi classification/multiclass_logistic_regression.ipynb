```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import load_digits # a dataset load_digits present in the sklearn data set library
```

```python
digits = load_digits()
```

```python
dir(digits)
```

```python
digits.target[0:5]
```

```python
# plt.gray() shows the image in black and white pixels
plt.gray()
plt.matshow(digits.images[15])
```

```python
for i in range(5):
    plt.matshow(digits.images[i])
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
X_train, X_test, y_train, y_test = tts(digits.data, digits.target, test_size = 0.1, random_state = 10)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
plt.matshow(digits.images[87])
```

```python
digits.target[87]
```

```python
model.predict([digits.data[87]])
```

```python
model.predict(digits.data[0:5])
```

```python
y_pre = model.predict(X_test)
from sklearn.metrics import confusion_matrix
```

```python
# this is the matrix which we wiil visualize in form of figure by using the seaborn library
cm = confusion_matrix(y_test, y_pre)
cm
```

```python
import seaborn as sn
plt.figure(figsize = (10,7)) # below figure size is decided by this or it will just us show us in default size
sn.heatmap(cm, annot = True) # this shows us the figure below
plt.xlabel('predicted')
plt.ylabel('truth')
```
