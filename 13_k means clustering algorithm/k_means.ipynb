 # Unsupervised learning 
 we dont have a target here we only got some features by which we will discover hidden patterns or data groupings in the datasets.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('income.csv')
```

```python
df
```

```python
plt.scatter(df['Age'], df['Income($)'], color = 'red')
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=3) # value of clustering meaning how many groups we wanna label the data
km
```

```python
y_pre = km.fit_predict(df[['Age', 'Income($)']]) # this performs clustering on df based on age and income and returns cluster labels(which category they belong to).
y_pre
```

```python
df['cluster'] = y_pre
df
```

```python
#here we divided the dataframe on the base of cluster they belong the cluster value is 3 so we divide them in three categories
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
plt.scatter(df1['Age'], df1['Income($)'], color = 'red')
plt.scatter(df2['Age'], df2['Income($)'], color = 'blue')
plt.scatter(df3['Age'], df3['Income($)'], color = 'green')
plt.xlabel('Age')
plt.ylabel('income')
plt.legend()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler() # we use this because as you can see in the above graph the scaling of x and y axis is not correct 
# and the plotting is not done correctly so we scale the values for age and income and then we plot them
scaler.fit(df[['Income($)']])
df['Income($)'] = scaler.transform(df[['Income($)']])
df
```

```python
scaler.fit(df[['Age']])
df['Age'] = scaler.transform(df[['Age']])
df
```

```python
km = KMeans(n_clusters= 3)
y_pre = km.fit_predict(df[['Age', 'Income($)']]) 
df.cluster = y_pre
df
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
plt.scatter(df1['Age'], df1['Income($)'], color = 'red')
plt.scatter(df2['Age'], df2['Income($)'], color = 'blue')
plt.scatter(df3['Age'], df3['Income($)'], color = 'green')
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], color = 'purple', marker ='*')
plt.xlabel('Age')
plt.ylabel('income')
plt.show()
```

```python
km.cluster_centers_ # this gives us the centriods of the clusters
```

```python
k_rng = range(1,10)
sse =[]
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df[['Age', 'Income($)']])
    sse.append(km.inertia_) # this km.intertie_ gives us the sum of squared errors, which we will further use in the elbow technique
```

```python
sse
```

```python
plt.plot(k_rng, sse)
plt.show()
```

```python
# in the above graph we use Elbow method which is a technique used to determine optimal number of k(n_clusters).
# we look for a elbow like structure in the graph and that point will be our k, like here k =3
# now we know we can use 3 clusters to discover hidden patterns or data groupings in the datasets.
```
