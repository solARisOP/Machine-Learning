```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
```

```python
df = pd.read_csv('sonar_dataset.csv', header = None)
df
```

```python
X = df.drop([60], axis =1)
y = df[60]
```

```python
y= y.replace({'R' : 1, 'M':0})
```

```python
y
```

```python
from sklearn.model_selection import train_test_split as tts
X_train, X_test, y_train, y_test = tts(X, y, test_size=0.25, random_state=5)
```

```python
X_train.shape, X_test.shape
```

```python
model = keras.Sequential([
    keras.layers.Dense(60, input_dim=60, activation = 'relu'),
    keras.layers.Dense(30,activation ='relu'),  
    keras.layers.Dense(15, activation = 'relu'),
    keras.layers.Dense(1, activation = 'sigmoid')
])

model.compile(optimizer = 'adam',
               loss='binary_crossentropy',
               metrics= ['accuracy'])

model.fit(X_train, y_train, epochs=100, batch_size = 8)
```

```python
model.evaluate(X_test, y_test)
```

```python
y_pred = model.predict(X_test).reshape(-1)
y_pred = np.round(y_pred)
y_pred
```

```python
from sklearn.metrics import confusion_matrix, classification_report
cm = confusion_matrix(y_test, y_pred)
print(classification_report(y_test, y_pred))
```

```python
model_d = keras.Sequential([
    keras.layers.Dense(60, input_dim=60, activation = 'relu'),
    keras.layers.Dropout(0.5), # it drops randomly certain amount of neurons in the hidden layers 0.5 means 50% neuron, this will help us in avoiding overfitting
    keras.layers.Dense(30,activation ='relu'),
    keras.layers.Dropout(0.5),    
    keras.layers.Dense(15, activation = 'relu'),
    keras.layers.Dropout(0.5),   
    keras.layers.Dense(1, activation = 'sigmoid')
])

model_d.compile(optimizer = 'adam',
               loss='binary_crossentropy',
               metrics= ['accuracy'])

model_d.fit(X_train, y_train, epochs=100, batch_size = 8)
```

```python
model_d.evaluate(X_test, y_test)
```

```python
y_pred_d = model_d.predict(X_test)
y_pred_d = np.round(y_pred_d)
```

```python
print(classification_report(y_test, y_pred_d))
```
