```python
import tensorflow_hub as hub
import tensorflow_text as text
```

```python
import tensorflow as tf
```

```python
preprocess_url = "https://tfhub.dev/tensorflow/bert_en_uncased_preprocess/3"
encoder_url = "https://tfhub.dev/tensorflow/bert_en_uncased_L-12_H-768_A-12/4"
```

preprocess our model

```python
bert_preprocess_model = hub.KerasLayer(preprocess_url)
```

```python
text_test = ['nice movie indeed', 'I love python programming']
text_preprocessed = bert_preprocess_model(text_test)
```

```python
text_preprocessed.keys()
```

```python
text_preprocessed['input_mask'] # five 1's for "nice movie indeed" is first 1 is for CLS and last 1 is for SEP and others in the middle are for the actual sentence and rest are the padding 
```

```python
text_preprocessed['input_word_ids']
```

```python
text_preprocessed['input_type_ids']
```

below will generate word/sentence embeddings

```python
bert_model = hub.KerasLayer(encoder_url)
bert_result = bert_model(text_preprocessed)
bert_result.keys()
```

## for any embedding may it for each word or each sentence, it a generates 768-length embedding vector

```python
bert_result['pooled_output'] # this is a embedding for each sentence it generates 768 outputs 
```

```python
bert_result['default'] # pooled is the default output which is for each sentence 
```

```python
bert_result['encoder_outputs'] # this is a embedding for each word in a sentence, there are 12 outputs coz of 12 encoder layers, each output is for each layer
```

```python
bert_result['sequence_output'] # final output for each word in a sentence 
```

```python
bert_result['encoder_outputs'][-1] == bert_result['sequence_output'] # the last encoder output is the final embedding vector output
```
