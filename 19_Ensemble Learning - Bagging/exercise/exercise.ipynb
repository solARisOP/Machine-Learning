```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('heart.csv')
```

```python
df.describe()
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
df[['Sex', 'ChestPainType', 'RestingECG', 'ExerciseAngina', 'ST_Slope']] = df[['Sex', 'ChestPainType', 'RestingECG', 'ExerciseAngina', 'ST_Slope']].apply(le.fit_transform)
```

```python
df
```

```python
df['zscore_old'] = (df.Oldpeak - df.Oldpeak.mean())/df.Oldpeak.std()
```

```python
df['zscore_max'] = (df.MaxHR - df.MaxHR.mean())/df.MaxHR.std()
```

```python
df['zscore_choles'] = (df.Cholesterol - df.Cholesterol.mean())/df.Cholesterol.std()
```

```python
df['zscore_resting_bp'] = (df.RestingBP - df.RestingBP.mean())/df.RestingBP.std()
```

```python
df['zscore_age'] = (df.Age - df.Age.mean())/df.Age.std()
```

```python
df
```

```python
df_new = df[((df.zscore_old > -3) &(df.zscore_old < 3)) & ((df.zscore_max > -3) &(df.zscore_max < 3)) & ((df.zscore_choles > -3) &(df.zscore_choles < 3)) & ((df.zscore_resting_bp > -3) &(df.zscore_resting_bp < 3))]
```

```python
df_new
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
```

```python
df_work = df_new.drop(['zscore_old', 'zscore_max', 'zscore_choles', 'zscore_resting_bp', 'zscore_age'],axis =1)
df_work
```

```python
y = df_work.HeartDisease
X = df_work.drop(['HeartDisease'], axis =1)
X_scale = scale.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split as tts
X_train, X_test, y_train, y_test = tts(X, y, stratify=y, random_state = 10)
```

```python
from sklearn.svm import SVC
model = SVC(C = 2200)
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
from sklearn.ensemble import BaggingClassifier
bg = BaggingClassifier(
                        base_estimator=model,
                        n_estimators= 100,
                        max_samples=0.8,
                        oob_score= True,
                        random_state=0)
bg.fit(X_train, y_train)
bg.oob_score_
```

```python
bg.score(X_test, y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
model_tree = DecisionTreeClassifier()
model_tree.fit(X_train, y_train)
model_tree.score(X_test, y_test)
```

```python
bg_tree = BaggingClassifier(
                        base_estimator=model_tree,
                        n_estimators= 100,
                        max_samples=0.8,
                        oob_score= True,
                        random_state=0)
bg_tree.fit(X_train, y_train)
bg_tree.oob_score_
```

```python
bg_tree.score(X_test, y_test)
```
