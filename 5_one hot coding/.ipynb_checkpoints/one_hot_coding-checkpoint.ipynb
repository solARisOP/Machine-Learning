```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('homeprices.csv')
```

```python
df.head()
```

```python
j = pd.get_dummies(df.town)
```

```python
merged = pd.concat([df, j], axis = 1)
```

```python
merged
```

```python
final = merged.drop(['town', 'west windsor'], axis = 1)
```

```python
final
```

```python
X = final.drop(['price'], axis =1)
```

```python
y = final[['price']]
```

```python
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
final
```

```python
model.predict([[2600, 1, 0]])
```
