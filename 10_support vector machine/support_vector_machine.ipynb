```python
import pandas as pd
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
dir(iris)
```

```python
df = pd.DataFrame(iris.data, columns = iris.feature_names)
```

```python
df['target'] = iris.target
iris.target_names
```

```python
df['flower'] = df.target.apply(lambda x : iris.target_names[x])
df.head()
```

```python
df0 = df[df.target == 0]
df1 = df[df.target == 1]
df2 = df[df.target == 2]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'], color = 'blue', marker= 'x')
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'], color = 'red', marker= 'x')
# plt.scatter(df2['sepal length (cm)'], df2['sepal width (cm)'], color = 'green', marker= 'x')
plt.show()
```

```python
X = df.drop(['flower', 'target'], axis =1)
```

```python
y = df.target
```

```python
from sklearn.model_selection import train_test_split as tts
X_train, X_test, y_train, y_test = tts(X,y, test_size = 0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.svm import SVC
model = SVC(gamma = 'auto', kernel = 'linear', C= 1000) # C(regulisation), gamma and kernel values helps in maintaining the accuracy of the model 
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
